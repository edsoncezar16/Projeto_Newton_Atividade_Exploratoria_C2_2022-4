```python
# convertendo os dados para uma tabela 'dataframe'
notas = []
cursos = []
turnos = []
for registro in dados:
    notas.append(registro.nota)
    cursos.append(registro.curso)
    turnos.append(registro.turno)
tabela_atividade = pd.DataFrame({
    'nota': notas, 'curso': cursos, 'turno': turnos
})
for questao in range(len(notas[0])):
    tabela_atividade[f'Q{questao + 1}'] = tabela_atividade.nota.apply(
        lambda nota: nota[questao]
    ) 
tabela_atividade.drop('nota', axis=1, inplace=True)
tabela_atividade = tabela_atividade.loc[:,[
    'Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'curso', 'turno'
]]
```

```python
tabela_atividade.columns = pd.Index(map(
    lambda label: label.upper(), tabela_atividade.columns
))
```

# Preparando as planilhas

Pontos por curso: "Cada linha é um estudante, e cada coluna indica a nota que ele tirou em cada questão"

Geral: "Cada linha contém a quantidade de estudantes de cada curso que tirou cada uma das pontuações totais"

```python
dados_c2 = pd.read_csv('dados_c2_limpos.csv')
dados_c2['Pont. Total'] = dados_c2.iloc[:,:7].apply(sum, axis=1)
dados_c2.head()
```

```python
dados_manha = dados_c2[dados_c2.TURNO == 'M']
dados_tarde = dados_c2[dados_c2.TURNO == 'T']
dados_NI = dados_c2[dados_c2.CURSO == 'Nao Identificado']
```

```python
def gerar_planilhas(dados, turno):
    pontos_por_curso = dados.groupby('CURSO')[
        dados.columns[:7].to_list() + ['Pont. Total']
    ]
    with pd.ExcelWriter(f'pontos_por_questao_{turno}.xlsx') as writer:
        for curso in pontos_por_curso.groups:
            pontos_por_curso.get_group(curso).to_excel(
                excel_writer=writer, sheet_name=curso, index=False
            )
    dados_gerais = dados.groupby(['CURSO', 'Pont. Total']).count()['Q1'].\
    to_frame().rename(columns={'Q1': 'No. de Estudantes'})\
    .reset_index(level=1).pivot(columns='Pont. Total').fillna(0).astype(int)
    dados_gerais.loc['Total por Nota'] = dados_gerais.sum(axis=0)
    dados_gerais['Total por Curso'] = dados_gerais.sum(axis=1)
    dados_gerais.to_excel(
        excel_writer=f'dados_gerais_{turno}.xlsx', sheet_name='Geral'
    )

```

```python
gerar_planilhas(dados_manha, 'manha')
```

```python
gerar_planilhas(dados_tarde, 'tarde')
```

```python
gerar_planilhas(dados_NI, 'nao-identificado')
```
