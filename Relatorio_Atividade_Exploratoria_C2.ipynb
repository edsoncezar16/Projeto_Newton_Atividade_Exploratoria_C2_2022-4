```python
# importanto bibliotecas 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid', rc={'figure.figsize':(12,6)})

# lendo os dados
dados_c2 = pd.read_csv('dados-ativ-expl-c2.csv', index_col=0)

# definindo funcoes  e variáveis adequadas
mapeamento_cursos = dict(enumerate(dados_c2.CURSO.unique()))

def analise_por_questao(dados, curso=None, mapeamento_cursos=mapeamento_cursos):
    if not curso is None:
        curso = mapeamento_cursos[curso]
        print(curso)
        dados_plot = dados[dados.CURSO == curso]
    else:
        dados_plot = dados
    print(f'Total de estudantes analisados: {dados_plot.shape[0]}')
    fig, axs = plt.subplots(2, 4)

    for n in range(7):
        ax = axs.flatten()[n]
        sns.histplot(
            data=dados_plot, x=f'Q{n + 1}', ax=ax,
            discrete=True, stat='percent', shrink=0.8
        )
        xticks = []
        for rectangle in ax.patches:
            prob = rectangle.get_height()
            x = rectangle.get_x() + rectangle.get_width() / 2
            y = rectangle.get_height()
            ax.annotate(f'{prob:.1f}', (x, y))
            #xticks.append(x)
        ax.set_xticks([0, 1, 2, 3])

    ax = axs.flatten()[-1]
    sns.histplot(
        data=dados_c2, x='Pont. Total', ax=ax,
        discrete=True, stat='percent', shrink=0.8
    )
    xticks = []
    for rectangle in ax.patches:
        prob = rectangle.get_height()
        x = rectangle.get_x() + rectangle.get_width() / 2
        y = rectangle.get_height()
        #ax.annotate(f'{prob:.1f}', (x, y))
        xticks.append(x)
    ax.set_xticks([0,7, 14, 21])
    fig.tight_layout();
    plt.show()
```

# Distribuição da pontuação por questões
